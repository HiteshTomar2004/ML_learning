```python
import numpy as np 
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

### Imports

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

### reading csv

```python
df = pd.read_csv("/kaggle/input/datasets/pablomgomez21/drugs-a-b-c-x-y-for-decision-trees/drug200.csv")
df.head()
```

### EDA

```python
df.info()
```

#### Label encoding every object in X
to convert categorical to numerical

```python
le_sex = LabelEncoder()
le_bp = LabelEncoder()
le_cholesterol = LabelEncoder()

df['Sex'] = le_sex.fit_transform(df['Sex'])
df['BP'] = le_bp.fit_transform(df['BP'])
df['Cholesterol'] = le_cholesterol.fit_transform(df['Cholesterol'])
```

Creating X and y for training

```python
X = df[['Age','Sex','BP','Cholesterol','Na_to_K']]
y = df['Drug']
```

train test split

```python
Xtrain,xtest,Ytrain,ytest = train_test_split(X,y,test_size=0.2,random_state=43)
```

## Decision Tree

```python
DT = DecisionTreeClassifier(random_state=20)
DT.fit(Xtrain,Ytrain)
y_prediction_DT = DT.predict(xtest)
```

## Bagging

```python
Bagg = BaggingClassifier(random_state=20,estimator=DecisionTreeClassifier(),n_estimators=50)
Bagg.fit(Xtrain,Ytrain)
y_prediction_Bagging = Bagg.predict(xtest)
```

## Random Forest

```python
RF = RandomForestClassifier(n_estimators=50,random_state=20)
RF.fit(Xtrain,Ytrain)
y_prediction_RF = RF.predict(xtest)
```

### evalutaion

```python
print('Accuracies:')
DT_accuracy = accuracy_score(ytest,y_prediction_DT)
Bagg_accuracy = accuracy_score(ytest,y_prediction_Bagging)
RF_accuracy = accuracy_score(ytest,y_prediction_RF)
print(f"Decision Tree accuracy: {DT_accuracy}")
print(f"Bagging accuracy: {Bagg_accuracy}")
print(f"Random Forest accuracy: {RF_accuracy}")
```

```python
print("Decision Tree:\n", classification_report(ytest, y_prediction_DT))
print("Bagging:\n", classification_report(ytest, y_prediction_Bagging))
print("Random Forest:\n", classification_report(ytest, y_prediction_RF))
```
